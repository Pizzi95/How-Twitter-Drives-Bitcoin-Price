<b>snscrape</b> is a scraper for social networking services (SNS). It scrapes things like user profiles, hashtags, or searches and returns the discovered items, e.g. the relevant posts. We used this service to analyze the last 4 months tweets about bitcoin and dogecoin.<br>
Compared to the Twitter api with snsscraper, we were able to analyse many more tweets and over a much longer period of time.

```python
import os
import pandas as pd
```

## QUERY BY USERNAME

```python
# Setting wanted username and number of tweets
tweet_count = 10000
username = "ElonMusk"

# Using OS library to call CLI commands in Python
os.system("snscrape --jsonl --max-results {} twitter-search 'from:{}'> user-tweets.json".format(tweet_count, username))
```

```python
# Reads the json generated from the CLI command above and creates a pandas dataframe
tweets_df1 = pd.read_json('user-tweets.json', lines=True)

# Displays first 5 entries from dataframe
tweets_df1.head()
```

## QUERY BY TEXT SEARCH

```python
# Setting variables to be used in format string command below
tweet_count = 300000
text_query = "#bitcoin"
since_date = "2021-01-01"
until_date = "2021-01-31"

# Using OS library to call CLI commands in Python
os.system('snscrape --jsonl --max-results {} --since {} twitter-search "{} until:{}"> text-query-tweets.json'.format(tweet_count, since_date, text_query, until_date))
```

```python
# Reads the json generated from the CLI command above and creates a pandas dataframe
tweets_df2 = pd.read_json('text-query-tweets.json', lines=True)

# Displays first 5 entries from dataframe
tweets_df2.head()
```
